### ANALISIS VARIABEL SELLING_PRICE

```python
import pandas as pd
import numpy as np
from scipy.stats import skew

data_selling_price = df['selling_price'].dropna()

mean = data_selling_price.mean()
median = data_selling_price.median()
modus = data_selling_price.mode()[0] if not data_selling_price.mode().empty else None

range_selling_price = data_selling_price.max() - data_selling_price.min()
variance = data_selling_price.var(ddof=0)  
std_dev = data_selling_price.std(ddof=0)  

skewness = skew(data_selling_price)

print("=== Analisis Pemusatan Data ===")
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
if modus is not None:
    print(f"Modus: {modus:.2f}")
else:
    print("Modus: Tidak ada nilai modus.")

print("\n=== Analisis Penyebaran Data ===")
print(f"Range: {range_selling_price:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")

print("\n=== Skewness (Kemiringan Data) ===")
print(f"Skewness: {skewness:.2f}")

if skewness > 0:
    print("Data miring ke kanan (Positif).")
elif skewness < 0:
    print("Data miring ke kiri (Negatif).")
else:
    print("Data simetris.")
```

Analisis statistik menunjukkan bahwa rata-rata harga jual adalah 300000,00 - 504127.31, dan mediannya 350000.00. Nilai skewness sebesar 4.89 menunjukkan distribusi dengan sedikit kemiringan ke kanan.
